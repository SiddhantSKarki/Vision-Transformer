## Model Architecture Code starts here

```python
n_embd = 32
input_channels = 1
img_size = 256
n_patches = 4
kernel_size = img_size // int((n_patches)**0.5)
out_channels = 1
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
class ViT(nn.Module):
    def __init__(self, imgs, gray_scale=False):
        super().__init__()
        # Convert an image into patches
        self.patches = self.convert_to_patches(imgs, gray_scale)
        
        
    @staticmethod
    def convert_to_patches(img, scale):
        if scale:
            # In case of gray-scaled images
            B, H, W = img.shape
            imgs = img.reshape(B, H // kernel_size, kernel_size, W // kernel_size, kernel_size)
            print(imgs.shape)
            return imgs.permute(0, 1, 3, 2, 4).reshape(-1, kernel_size, kernel_size)
            
        # In case of colored images
        B, H, W, C = img.shape
        imgs = img.reshape(B, H // kernel_size, kernel_size, W // kernel_size, kernel_size, C)
        return imgs.permute(0, 1, 3, 2, 4).reshape(-1, kernel_size, kernel_size, C)
```

```python
image = torch.Tensor(test_image[None, ...])
model = ViT(image, gray_scale=True)
```
